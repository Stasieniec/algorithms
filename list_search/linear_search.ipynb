# Linear search

**Pretty basic stuff**

```python
from typing import List
```

```python
def linear_search(l: List[any], element: any) -> int:
    """
    Finds an element in the list

    :param l: the list
    :param element: the element
    :returns: index of the element.
    """
    for i, x in enumerate(l):
        if x == element:
            return i

    return -1
```

```python
l = [1, 3, 5, 23, 6, 325, 435, 232]
index: int = linear_search(l, 3)
print(f'Tis element is at the index of {index}')
```
