# Deep First Search
**An example of BFS on a simple tree with use of dictionaries.**

```python
from typing import List, Dict
```

```python
edges: Dict[int, List[int]] ={
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 9, 8, 6],
    3: [1],
    4: [1, 7],
    5: [10],
    6: [2],
    7: [4],
    8: [2],
    9: [2, 10],
    10: [5, 9],
    11: [12],
    12: [11]

}
```

```python
def dfs(node: int, target: int, stack: List[int], visited: List[bool]) -> bool:
    """
    Checks if a node is reachable from another node

    :param node: current node
    :param target: goal node
    :stack: current state of the stack
    :visited: list of booleans representing visited nodes
    :returns: True or False, accordingly to whether the target node is reachable.
    """
    print(f'I am in the node {node}')
    if node == target:
        return True
    visited[node] = True
    for x in edges[node]:
        if visited[x] == False:
            stack.append(x)
    
    if len(stack) > 0:
        return dfs(stack.pop(), target, stack, visited)
    else:
        return False
```

```python
dfs(0, 11, [], [False]*13)
```
