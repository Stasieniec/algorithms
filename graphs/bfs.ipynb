# Best First Search

**An example of BFS on a simple tree with use of dictionaries.**

```python
from  typing import List, Dict, Tuple
```

**Making a tree with edges**

```python
edges: Dict[int, List[int]] ={
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 9, 8, 6],
    3: [1],
    4: [1, 7],
    5: [10],
    6: [2],
    7: [4],
    8: [2],
    9: [2, 10],
    10: [5, 9],
    11: [12],
    12: [11]

}
```

**Algorithm** 

The purpose of the algorithm is to see if one node can be reached from another.

```python
def bfs(node: int, target: int) -> bool:
    visited: List[bool] = [False]*13
    queue: List[int] = [node]
    while queue:
        x = queue.pop(0)
        print(f'I am now in the node {x}')
        visited[x] = True
        if x == target:
            return True

        for adjancted in edges[x]:
            if visited[adjancted] == False:
                queue.append(adjancted)

    return False

    
```

```python
bfs(0, 12)
```
