# Dijkstra's algorithm
**Finding the shortest path from start node to every other node in a graph**

In this example the graph is weighted and directed.

```python
# import
from typing import List, Dict
```

Create a weighted, directed graph:

```python
nodes: List[str] = ['A', 'B', 'C',' D', 'E', 'F']

graph: List[int] = [
#    A  B  C  D  E  F
    [0, 1, 0, 0, 0, 0], #A
    [1, 0, 2, 0, 0, 0], #B
    [0, 0, 0, 1, 0, 0], #C
    [0, 0, 0, 0, 4, 0], #D
    [0, 1, 0, 0, 0, 1], #E
    [0, 0, 7, 0, 0, 0] #F
]

# graph[i][j] = x means that there is an edge from i to j with weight x.
# if the weight is 0, then the edge does not exist.
```

```python
def dijkstra(graph: List[int], start: int, current: int, visited: List[int] = [], distances: List[int] = [9999]*len(graph)) -> List[int]:
    """
    Calculates the minimal distance from the start node to 
    each node in a graph.

    :param graph: the graph
    :param start: starting node
    :param current: current node
    :param visited: list of visited nodes
    :returns: list of distances from the start node.
    """
    distances[start] = 0
    while len(visited) != len(graph):
        curr_node: int = 0
        curr_dist: int = 999
        for i, distance in enumerate(distances):
            if (i not in visited) and (distance < curr_dist):
                curr_dist = distance
                curr_node = i

        for i, edge in enumerate(graph[curr_node]):
            distances[i] = min(edge + distance, distances[i])

        visited.append(curr_node)
        


    return distances
```

```python
dijkstra(graph, 0, 0)
```
